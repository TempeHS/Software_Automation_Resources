__4. In a markdown block, outline the Machine Learning Linear Regression algorithm.__

The machine learning linear regression algorithm is an algorithm that predicts values based on data it has been trained on. It will be given a dataset with two variables, and will create a linear graph between these two variables to predict values of one variable from the other. This linear graph follows the equation y = wx + b where w is the weight, effecting the gradient of the line, and b is the bias, moving the y-intercept of the line up or down. Changing these two values, and using a cost function, the algorithm finds the closest/best fit to the data, or where the derivative of the mean squared cost is 0.
